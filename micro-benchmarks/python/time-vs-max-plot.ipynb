```python
def temp_to_time(dict1, dict2):
    metric_dict1 = {}
    metric_dict2 = {}
    for temp, time in dict1.items():
        # Find the minimum time of a temperature
        if temp not in metric_dict1 or metric_dict1[temp] > time:
            metric_dict1[temp] = time
    for temp, time in dict2.items():
        # Find the minimum time of a temperature
        if temp not in metric_dict2 or metric_dict2[temp] > time:
            metric_dict2[temp] = time
    
    fig=plt.figure()
    ax=fig.add_subplot(111, label="1")
    ax2=fig.add_subplot(111, label="2", frame_on=False)
    
    ax.plot(dict1.values(), dict1.keys(), color="C0")
    ax.set_xlabel("time1", color="C0")
    ax.set_ylabel("temp1", color="C0")
    ax.tick_params(axis='x', colors="C0")
    ax.tick_params(axis='y', colors="C0")
    ax.set_xlim(20, 260)
    ax.set_ylim(70, 100)

    
    ax2.plot(dict2.values(), dict2.keys(), color="C1")
    ax2.xaxis.tick_top()
    ax2.yaxis.tick_right()
    ax2.set_xlabel('time2', color="C1") 
    ax2.set_ylabel('temp2', color="C1")       
    ax2.xaxis.set_label_position('top') 
    ax2.yaxis.set_label_position('right') 
    ax2.tick_params(axis='x', colors="C1")
    ax2.tick_params(axis='y', colors="C1")
    ax.set_xlim(20, 260)
    ax.set_ylim(70, 100)

    
    # draw vertical line from (70,100) to (70, 250)
    ax.plot([13, 260], [75, 75], 'k-', lw=1, color="r")

    plt.show()
```

```python
# Numerical Differentiation
def num_diff(X, Y):
    Y_diff = []
    for i, _ in enumerate(X, start=1):
        if i < len(X):
            Y_diff.append((Y[i] - Y[i - 1]) / X[i - 1])
    return Y_diff, X[1:]
    # plt.plot(X[1:], Y_diff)
```

```python
def plot_two_diff(df1, header1, header2, df2, header3, header4):
    fig=plt.figure()
    
    ax=fig.add_subplot(111, label="1")
    ax2=fig.add_subplot(111, label="2", frame_on=False)
    ax.set_title("Tensorflow-Matmul-Differentiation", y=1.2)
    
    Y1_diff, X1 = num_diff(np.log(df1[header1]), np.log(df1[header2]))
    ax.plot(X1, Y1_diff, color="C0")
    ax.set_xlabel("time", color="C0")
    ax.set_ylabel("Max_temp_diff", color="C0")
    ax.tick_params(axis='x', colors="C0")
    ax.tick_params(axis='y', colors="C0")
    ax.set_xlim(2.8, 4.3)
    ax.set_ylim(-0.005, 0.023)
    
    # draw horizontal line
    ax.plot([2.8, 4.3], [mean(Y1_diff), mean(Y1_diff)], 'k-', lw=1, color="C0")
    print (mean(Y1_diff))
    
    Y2_diff, X2 = num_diff(np.log(df2[header3]), np.log(df2[header4]))
    ax2.plot(X2, Y2_diff, color="C1")
    ax2.xaxis.tick_top()
    ax2.yaxis.tick_right()
    ax2.set_xlabel('time', color="C1") 
    ax2.set_ylabel("Max_temp_diff", color="C1")       
    ax2.xaxis.set_label_position('top') 
    ax2.yaxis.set_label_position('right') 
    ax2.tick_params(axis='x', colors="C1")
    ax2.tick_params(axis='y', colors="C1")
    ax2.set_xlim(2.8, 4.3)
    ax2.set_ylim(-0.005, 0.023)

    ax2.plot([2.8, 4.3], [mean(Y2_diff), mean(Y2_diff)], 'k-', lw=1, color="C1")
    print (mean(Y2_diff))
    
    plt.show()
```
